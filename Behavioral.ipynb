```
def get_avg_hashtags(strings):
    count = 0
    for s in strings:
        count = count + len(set(part[1:] for part in str(s).split(' ') if part.startswith('#')))
    return count / len(strings)
```

```
def get_avg_hashtag_chars(strings):
    count = 0
    for s in strings:
        #print(s)
        ls = list(set(part[1:] for part in str(s).split(' ') if part.startswith('#')))
        #print(ls)
        for element in ls:
            count = count + len(element)
    return count / len(strings)
```

```
def get_avg_mentions(strings):
    count = 0
    for s in strings:
        count = count + len(set(part[1:] for part in str(s).split(' ') if part.startswith('@') and len(part) > 2))
    return count / len(strings)
```

```
def get_avg_links(strings):
    count = 0
    for s in strings:
        length = len(list(re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', str(s))))
        count = count + length
    return count / len(strings)
```

```
def contain_links(strings):
    count = 0
    for s in strings:
        length = len(list(re.findall('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', str(s))))
        if length > 0:
            count = count + 1
    return count / len(strings)
```

```
def get_total_tweets(strings):
    return len(strings)
```

```
def get_tot_days(strings):
    return len(set(strings))
```

```
def get_retweet_rate(numbers):
    return (np.count_nonzero(numbers)) / len(numbers)
```

```
def get_avg_tweet_length(strings):
    return np.mean([len(str(s)) for s in strings])
```
